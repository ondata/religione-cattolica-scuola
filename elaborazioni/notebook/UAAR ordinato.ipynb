```python
import pandas as pd
```

```python
data_2019 = pd.read_excel('Alunni che frequentano Rel Catt_2018-19_2019-20_2020-21.xlsx', sheet_name=0)
data_2020 = pd.read_excel('Alunni che frequentano Rel Catt_2018-19_2019-20_2020-21.xlsx', sheet_name=1)
data_2021 = pd.read_excel('Alunni che frequentano Rel Catt_2018-19_2019-20_2020-21.xlsx', sheet_name=2)
```

```python
data_2019 = data_2019.set_index('Codice Scuola')
data_2020 = data_2020.set_index('Codice Scuola')
data_2021 = data_2021.set_index('Codice Scuola')
```

```python
data = data_2019.join(data_2020, how='outer')
data = data.join(data_2021, how='outer')
```

```python
data
```

```python
statali_2021 = pd.read_csv("SCUANAGRAFESTAT20202120210831.csv")
statali_2020 = pd.read_csv("SCUANAGRAFESTAT20192020200831.csv")
statali_2019 = pd.read_csv("SCUANAGRAFESTAT20181920190831.csv")
```

```python
statali_2019 = statali_2019.set_index('CODICESCUOLA')
statali_2020 = statali_2020.set_index('CODICESCUOLA')
statali_2021 = statali_2021.set_index('CODICESCUOLA')
```

```python
paritarie_2019 = pd.read_csv("SCUANAGRAFEPAR20181920190831.csv") 
paritarie_2020 = pd.read_csv("SCUANAGRAFEPAR20192020200831.csv") 
paritarie_2021 = pd.read_csv("SCUANAGRAFEPAR20202120210831.csv") 
```

```python
paritarie_2019 = paritarie_2019.set_index('CODICESCUOLA')
paritarie_2020 = paritarie_2020.set_index('CODICESCUOLA')
paritarie_2021 = paritarie_2021.set_index('CODICESCUOLA')
```

```python
scuole = pd.concat([statali_2019, statali_2020, statali_2021, paritarie_2019, paritarie_2020, paritarie_2021])
```

```python
scuole
```

```python
data = data.join(scuole, how='left')
```

```python
data
```

```python
data = (data.reset_index()
        .drop_duplicates(subset='index', keep='last')
        .set_index('index').sort_index())
```

```python
data = data.dropna(how='all')
```

```python
data = data.rename(columns={"Numero di alunni che frequentano l'ora di religione - a.s. 2018/19": '2018/19', "Numero di alunni che frequentano l'ora di religione - a.s. 2019/20": '2019/20', "Numero di alunni che frequentano l'ora di religione - a.s. 2020/21": '2020/21'})
```

```python
data['2020/21'] = data['2020/21'].replace(['<=3'],'3')
data['2019/20'] = data['2019/20'].replace(['<=3'],'3')
data['2018/19'] = data['2018/19'].replace(['<=3'],'3')
```

```python
data = data.fillna(0)
```

```python
data["2018/19"] = pd.to_numeric(data["2018/19"])
data["2019/20"] = pd.to_numeric(data["2019/20"])
data["2020/21"] = pd.to_numeric(data["2020/21"])
```
